What is the file function in Python?

In Python, you work with files using the built-in open() function.
This function opens a file and returns a file object, which allows you to read, write, or modify the file.

Syntax
file_object = open("filename", "mode")

Keywords / Modes used to create and write files

Python uses file modes (keywords) to specify what you want to do with the file:

To Create a File

x → creates a new file; gives an error if file exists

w → creates a file if it does not exist, overwrites if it does

To Write to a File

w → write mode (overwrite file)

a → append mode (add new data to the end of file)

```python
#78) Write a python program to find the longest words. 
# File name: sample.txt

longest_words = []
max_length = 0

with open(r"1.txt",) as file:
    for line in file:
        words = line.split()
        for word in words:
            length = len(word)
            
            if length > max_length:
                max_length = length
                longest_words = [word]       
            
            elif length == max_length:
                longest_words.append(word)   

print("Longest word length:", max_length)
print("Longest word(s):", longest_words)

```

```python

#80) Write a Python program to count the frequency of words in a file. 

word_count = {}   

with open("1.txt", "r") as file:
    for line in file:
        words = line.split()             
        for word in words:
            word = word.lower()         
            if word in word_count:
                word_count[word] += 1     
            else:
                word_count[word] = 1      

print("Word Frequency in File:")
for word, count in word_count.items():
    print(word, ":", count)

```

```python
#90) Write python program that user to enter only odd numbers, else 
#will raise an exception.


class NotOddNumberError(Exception):
    pass

try:
   
    num = int(input("Enter an odd number: "))
    
    if num % 2 == 0:
        raise NotOddNumberError("Error: You must enter an odd number!")
    else:
        print(f"You entered  odd number: {num}")

except ValueError:
    print("Error: Please enter a valid number.")

except NotOddNumberError as e:
    print(e)
```
